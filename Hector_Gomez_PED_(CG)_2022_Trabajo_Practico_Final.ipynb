##Ejercicio 2

Defina otro dataframe cumpliendo lo siguiente: 

*   Imputar los datos faltantes y valores atípicos, descartar el resto. 

*   Reemplazar los valores de las columnas no numéricas por valores numéricos. (Por ejemplo si es mujer poner 1, si es varón poner 0)

Observacion: Un valor atípico es aquel que dista mas de 2 desvios standard para el contexto de este ejercicio.

```python
#Verifico los datos faltantes de los dataset

# El total de datos nulos en la tabla
df.isnull().sum()
```

```python
#Cambio los datos de sexos en números
# Mujer = 1 y Hombre = 0
df['Sex'].replace('female',1,inplace=True)
df['Sex'].replace('male',0,inplace=True)
```

```python
#Cambio los datos de embarque en números
# Puerto de embarque (C = Cherburgo, Q = Queenstown, S = Southampton)
# C = 1, Q = 2 y S = 3
df['Embarked'].replace('C',1,inplace=True)
df['Embarked'].replace('Q',2,inplace=True)
df['Embarked'].replace('S',3,inplace=True)
```

```python
#Reemplazo los datos faltantes en la tarifa

#Decidi imputar los datos de la columna Fare remplazando el unico valor faltante por la media de todos los demas valores.
df['Fare'].interpolate(method ='linear', limit_direction = "forward",inplace=True)
#Verifico que se cambio el dato Nulo
df.isnull().sum()
df.describe()
```

```python
#Reemplazo los datos faltantes en la edad 

# df.corr(method='pearson')
#No existe ninguna correlacion fuerte de la columna Age con los demas datos, lo que nos inpide hacer una imputacion por corelacion

#Decidi Interpolar los datos de la columna [Age]
df['Age'].interpolate(method ='linear', limit_direction = "both",inplace=True)

# df.describe()
df.isnull().sum()
```

```python
#Si considero necesario eliminar alguna columna
df.isnull().sum()
# Sabiendo que en la Columna [Cabin] existen 1014 datos faltantes de un total de 1309 en nuestro DataFrame.
# Es una columna que considerare eliminar por su cantidad de datos Nulos y su poca relevancia en este analisis.
df.drop(columns=['Cabin'], axis=1, inplace=True)
df
```

```python
#Elimino las columnas que considero que no son necesarias para el analisis

# Considero que la Columna Name y la Columna Ticket no son relevantes para nuestro analisis
# Son datos categoricos caracter que no tienen alguna correlacion con algun otro dato.

df.drop(columns=['Name'], axis=1, inplace=True)
df.drop(columns=['Ticket'], axis=1, inplace=True)
```

*Mala Imputacion de datos*

Las malas imputaciones pueden llevar tener graficos con distribuciones raras y por lo que nos puede llevar a una mala lectura de nuestros datos, fallando en nuestro analisis. 

De haber sustituido los datos Nulos de la columna [Age] por 0 , hubiesemos tenido una cantidad de datos que no nos indican nada y nos cambiarian la distribucion original de los datos. Originalmente el histograma de edad nos mostraba una distribucion normal (en forma de campana) de haber cambiado los datos por 0 hubiesemos perdido esa distribucion con una barra en 0 de mas de 230 pasajeros.
En nuestro caso hicimos una interpolacion linear a ambos lados de arriba y abajo donde no tuvimos una gran diferencia en la distribucion de los datos de nuestro histograma.
